```python
from sympy import *
from toolbox import Tools

tools = Tools()
V_POC, I11, I12, I21, I22, I31, I32, P1, P2, V_A1, V_A2, V_B, V_S1, V_S2, Z11, Z12, Z21, Z22, Z31, Z32, Z41, Z42, Z5, Z6 = symbols('V_POC I11 I12 I21 I22 I31 I32 P1 P2 V_A1 V_A2 V_B V_S1 V_S2 Z11 Z12 Z21 Z22 Z31 Z32 Z41 Z42 Z5 Z6')
Z11, Z12, Z21, Z22, Z31, Z32, Z41, Z42, Z5, Z6 = tools.get_impedances(verb=True)
init_printing(use_unicode=True)
```

# Equivalent impedances

### Turbine 1

```python
zx = (Z31**-1 + Z11**-1)**-1
zy = Z21 + zx
ZeqT1 = (Z41**-1 + zy**-1)**-1
ZeqT1
```

```python
zx = (Z32**-1 + Z12**-1)**-1
zy = Z22 + zx
ZeqT2 = (Z42**-1 + zy**-1)**-1
ZeqT2
```

```python
ZeqPOC = Z5
ZeqPOC
```

```python
ZT1 = Z11 + (Z31**-1 + (Z21 + (ZeqT2**-1 + ZeqPOC**-1 + Z41**-1)**-1)**-1)**-1
```

```python
solve(V_S1**2 / ZT1 - 3e6, V_S1)
```

# Node Equations

```python

eq1 = (V_S1 - V_A1) / Z11 + V_A1 / Z31 + (V_A1-V_B) / Z21
eq2 = (V_A1 - V_B) / Z21 + V_B / (Z41**-1 + ZeqT2**-1 + ZeqPOC**-1)**-1
sol = solve([eq1, eq2], [V_A1, V_B])
IT1 = list(sol.values())
IT1
```

# Mesh equations
### Turbine 1

```python
i_A, i_B, i_C = symbols('i_A i_B i_C')
eqA = -V_S1 + i_A * Z11 + (i_A - i_B) * Z31
#eqA = eqA.subs('i_A', P1 / V_S1)
eqA
```

```python
eqB = (i_B - i_A) * Z31 + i_B * Z21 + (i_B - i_C) * Z41
eqB
```

```python
eqC = (i_C - i_B) * Z41 + i_C * (ZeqT2**-1 + ZeqPOC**-1)**-1
eqC
```

```python
sol = solve([eqA, eqB, eqC], [i_A, i_B, i_C])
i_T1 = list(sol.values())
i_T1
```

```python

VB_T1 = (i_T1[1] - i_T1[2]) * Z41
VB_T1
```

```python
eq1 = VB_T1.simplify()
```

### Turbine 2

```python
i_A, i_B, i_C = symbols('i_A i_B i_C')
eqA = -V_S2 + i_A * Z12 + (i_A - i_B) * Z32
eqB = (i_B - i_A) * Z32 + i_B * Z22 + (i_B - i_C) * Z42
eqC = (i_C - i_B) * Z42 + i_C * (ZeqT1**-1 + ZeqPOC**-1)**-1

sol = solve([eqA, eqB, eqC], [i_A, i_B, i_C])
i_T2 = list(sol.values())
VB_T2 = (i_T2[1] - i_T2[2]) * Z42
VB_T2
```

### POC

```python
i_A, i_B, i_C = symbols('i_A i_B i_C')
eqA = -V_POC + (i_A - i_B) * Z6
eqB = (i_B - i_A) * Z6 + i_B * Z5 + i_B * (ZeqT1**-1 + ZeqT2**-1)**-1

sol = solve([eqA, eqB], [i_A, i_B])
i_POC = list(sol.values())
i_POC
```

```python
VB_POC = V_POC - i_POC[1] * Z5
VB_POC
```

```python
 Z11, Z12, Z21, Z22, Z31, Z32, Z41, Z42, Z5, Z6 = tools.get_impedances(verb=True)
```

```python
eq_POC = VB_POC.subs({'Z11':Z11, 'Z12':Z12, 'Z21':Z21, 'Z22':Z22, 'Z31':Z31, 'Z32':Z32, 'Z41':Z41, 'Z42':Z42, 'Z5':Z5, 'Z6':Z6, 'V_POC':4e3}).simplify()

eq_T1 = VB_T1.subs({'Z11':Z11, 'Z12':Z12, 'Z21':Z21, 'Z22':Z22, 'Z31':Z31, 'Z32':Z32, 'Z41':Z41, 'Z42':Z42, 'Z5':Z5, 'Z6':Z6}).simplify()

eq_T2 = VB_T2.subs({'Z11':Z11, 'Z12':Z12, 'Z21':Z21, 'Z22':Z22, 'Z31':Z31, 'Z32':Z32, 'Z41':Z41, 'Z42':Z42, 'Z5':Z5, 'Z6':Z6}).simplify()
```
